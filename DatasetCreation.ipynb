# Dataset Creation

Images of human emotions (happy, sad, angry and neutral) were scraped from Bing and saved into corresponding folders. These folders are located in the dataset directory. Each label has 500-600 images. 

Two tools were used for web scraping:


1.   https://github.com/ostrolucky/Bulk-Bing-Image-downloader
- This scraping tool does not download duplicates and allows us to apply filters on the search. Our team used this to make sure all images are unique. However, it only downloads a small number of images at a time.
2.   https://github.com/gurugaurav/bing_image_downloader
- This tool allows us to set the number of images to be downloaded.



# Downloading images

The process is done locally using command prompt.

Here is the code for downloading images for "happy" emotion.

```python
# Using tool 1
./bbid.py -s "happy emotion" -o "happy" --filters +filterui:face-face+filterui:photo-photo
```

The argument for filters: "+filterui:face-face+filterui:photo-photo" specfies that we only want images of human faces. This allows us to eliminate irrelevant photos, such as emojis, cartoon images, etc.

```python
# Using tool 2
# installing
pip install bing-image-downloader
```

```python
# using command prompt to cd into the directory of the project first and then start coding
from bing-image-downloader import downloader
downloader.download("happy emotion", limit=600, output_dir='dataset', adult_filter_off=True, force_replace=False, timeout=180)
```

We repeated the process with a different argument for -s in tool 1 and the first argument in tool 2 (the query string). Specifically, we changed the query string to: happy face person, happy expression, smiling face, happy mood, etc. Trying different queries allowed us to find more unique images. 

We then combined the results of two tools and deleted duplicated or irrelevant images.

The same process was done to scrape images for sad, angry and neutral emotions.

```python
# scraping images for "sad" emotion
# tool 1
./bbid.py -s "sad emotion" -o "sad" --filters +filterui:face-face+filterui:photo-photo
# tool 2
downloader.download("sad emotion", limit=600, output_dir-'dataset', adult_filter_off=True, force_replace=False, timeout=180)
# the query string was then changed to: sad face person, sad expression, sad mood, etc
```

```python
# scraping images for "angry" emotion
# tool 1
./bbid.py -s "angry emotion" -o "angry" --filters +filterui:face-face+filterui:photo-photo
# tool 2
downloader.download("angry emotion", limit=600, output_dir='dataset', adult_filter_off=True, force_replace=False, timeout=180)
# the query string was replace with: angry expression, angry face, angry face person, angry mood, etc
```

```python
# scraping images for "neutral" emotion
# tool 1
./bbid.py -s "neutral emotion" -o "neutral" --filters +filterui:face-face+filterui:photo-photo
# tool 2
downloader.download("neutral emotion", limit=600, output_dir='dataset', adult_filter_off=True, force_replace=False, timeout=180)
# the query string was replace with: neutral expression, neutral face, neutral face person, blank expression, etc
```
